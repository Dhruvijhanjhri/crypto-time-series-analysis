```python
import pandas as pd
```

```python
df = pd.read_csv("crypto_master_data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
master_df = df[
    ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'crypto']
]
```

```python
master_df.rename(columns={
    'Open': 'open_price',
    'High': 'high_price',
    'Low': 'low_price',
    'Close': 'close_price',
    'Volume': 'volume'
}, inplace=True)
```

```python
master_df.sort_values(by=['crypto', 'Date'], inplace=True)
```

```python
master_df.set_index('Date', inplace=True)
```

```python
master_df.head()
```

```python
master_df.groupby('crypto').size()
```

```python
master_df.to_csv("crypto_master_cleaned.csv")
```
