```python
import os
import pandas as pd

print("üìç Working directory:", os.getcwd())

DATA_DIR = os.path.join("..", "data", "processed")

ARIMA_PATH   = os.path.join(DATA_DIR, "arima_all_coins.csv")
SARIMA_PATH  = os.path.join(DATA_DIR, "sarima_all_coins.csv")
PROPHET_PATH = os.path.join(DATA_DIR, "prophet_all_coins.csv")
LSTM_PATH    = os.path.join(DATA_DIR, "lstm_all_coins.csv")
LIVE_PATH    = os.path.join(DATA_DIR, "live_crypto_data.csv")

FINAL_PATH   = os.path.join(DATA_DIR, "final_crypto_forecast_master.csv")
```

```python
import pandas as pd

arima_df   = pd.read_csv("arima_all_coins.csv")
sarima_df  = pd.read_csv("sarima_all_coins.csv")
prophet_df = pd.read_csv("prophet_all_coins.csv")
lstm_df    = pd.read_csv("lstm_all_coins.csv")
live_df    = pd.read_csv("live_crypto_data.csv")

print("‚úÖ All forecast files loaded")
```

```python
def clean(df):
    df.columns = df.columns.str.lower()
    df["date"] = pd.to_datetime(df["date"])
    return df[["date", "crypto", "forecast_price", "model"]]

arima_df   = clean(arima_df)
sarima_df  = clean(sarima_df)
prophet_df = clean(prophet_df)
lstm_df    = clean(lstm_df)
live_df    = clean(live_df)

print("‚úÖ Columns standardized")
```

```python
prophet_df = prophet_df.rename(columns={
    "ds": "date",
    "yhat": "forecast_price"
})

prophet_df["crypto"] = "BTC"
prophet_df["model"] = "PROPHET"
prophet_df = prophet_df[["date", "crypto", "forecast_price", "model"]]
```

```python
final_df = pd.concat(
    [arima_df, sarima_df, prophet_df, lstm_df, live_df],
    ignore_index=True
)

final_df.sort_values(["crypto", "date", "model"], inplace=True)
final_df.head()
```

```python
print("Coins:", final_df["crypto"].unique())
print("Models:", final_df["model"].unique())
print("Missing values:\n", final_df.isna().sum())
print("Total rows:", len(final_df))
```

```python
final_df.to_csv("final_crypto_forecast_master.csv", index=False)
print("‚úÖ final_crypto_forecast_master.csv created")
```
