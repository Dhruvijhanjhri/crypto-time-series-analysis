```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arima_sarima_df = pd.read_csv("arima_sarima_forecasts.csv")
prophet_df = pd.read_csv("prophet_btc_forecast.csv")
lstm_df = pd.read_csv("lstm_btc_forecast.csv")
live_df = pd.read_csv("live_crypto_data.csv")
```

```python
print("ARIMA/SARIMA:", arima_sarima_df.columns)
print("PROPHET:", prophet_df.columns)
print("LSTM:", lstm_df.columns)
print("LIVE:", live_df.columns)
```

```python
arima_sarima_df.columns = arima_sarima_df.columns.str.lower()

arima_sarima_df = arima_sarima_df.rename(
    columns={'forecast': 'forecast_price'}
)

arima_sarima_df = arima_sarima_df[
    ['date', 'crypto', 'forecast_price', 'model']
]
```

```python
prophet_df.columns = prophet_df.columns.str.lower()

prophet_df = prophet_df.rename(columns={
    'ds': 'date',
    'yhat': 'forecast_price'
})

prophet_df['crypto'] = 'BTC'
prophet_df['model'] = 'PROPHET'

prophet_df = prophet_df[
    ['date', 'crypto', 'forecast_price', 'model']
]
```

```python
lstm_df.columns = lstm_df.columns.str.lower()

lstm_df = lstm_df.rename(columns={
    'prediction': 'forecast_price'
})

lstm_df['crypto'] = 'BTC'
lstm_df['model'] = 'LSTM'

lstm_df = lstm_df[
    ['date', 'crypto', 'forecast_price', 'model']
]
```

```python
live_df.columns = live_df.columns.str.lower()

live_df = live_df[
    ['date', 'crypto', 'forecast_price', 'model']
]
```

```python
final_df = pd.concat(
    [arima_sarima_df, prophet_df, lstm_df, live_df],
    ignore_index=True
)

final_df['date'] = pd.to_datetime(final_df['date'])
final_df.sort_values(['crypto', 'date', 'model'], inplace=True)

final_df.head()
```

```python
final_df.to_csv("final_crypto_forecast_master.csv", index=False)
```
